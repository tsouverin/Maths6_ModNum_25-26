# **Exercice 2.A — Limite de Roche (7 points)**

Lorsque deux corps orbitent l'un autour de l'autre, il sont soumis à des forces qui les déforment. On les appelle les **forces de marée**, et elles sont fonction de la distance qui sépare les deux corps. C'est de cette façon que notre satellite, la Lune, déforme les océans et les mers de la Terre, générant ainsi le **phénomène des marées** que l'on connaît. Ces forces de marée sont d'autant plus fortes que le satellite est proche de la planète. Ainsi, si un satellite s'approche trop près de la planète, les forces de marée peuvent devenir si fortes qu'elles peuvent **disloquer** le satellite en morceaux. 

La distance à laquelle cela se produit est appelée la **limite de Roche**, telle que :

$$
d_\mathrm{Roche} \simeq 2.44 \times R_p \left(\frac{\rho_p}{\rho_s} \right)^{\frac{1}{3}} \qquad (A.1)
$$

avec :

- $d_\mathrm{Roche}$ : distance minimale pour éviter la dislocation (en mètres).
- $R_p$ : rayon de la planète (en mètres).
- $\rho_p$​ : masse volumique de la planète (en kg/m³).
- $\rho_s$​ : masse volumique du satellite (en kg/m³).

<center>

| Corps céleste                   | Masse volumique (en kg/m³)| Distance moyenne à la planète (m) | Rayon (m) |
|---------------------------------|---------------------------|-----------------------------------|-----------|
| Terre                           | 5515                      | -                                 |6.38 × 10⁶ |
| Lune                            | 3346                      | 3.844 × 10⁸                       |1.74 × 10⁶ |
| Jupiter                         | 1328                      | -                                 |7.15  × 10⁷|
| Europe (satellite de Jupiter)   | 3010                      | 6.71 × 10⁸                        |1.561 × 10⁶|
| Io (satellite de Jupiter)       | 3528                      | 4.22 × 10⁸                        |1.822 × 10⁶|
| Adrastée (satellite de Jupiter) | 860                       | 1.29 × 10⁸                        |1.6 × 10⁴  |

*Tableau A.1 — Paramètres physiques de différents corps célestes.*

</center>

**1. Calculez la limite de Roche pour la Terre et la Lune à l'aide de l'équation (A.1). (1 point)**

**2. En admettant qu'un satellite est à l'abri de la dislocation s'il est à une distance supérieure à $1.1 \times d_\mathrm{Roche}$, comparez cette limite à la distance moyenne entre la Terre et la Lune et concluez sur le fait que la Lune est ou non à l'abri d'une dislocation. (1 points)**

**3. Faîtes de même pour Europe, Io et Adrastée, tous satellites de Jupiter. (1.5 points)**

**4. On sait que la Lune s'éloigne de la Terre à une vitesse de 0.038 m par an. Imaginons que l'inverse se produise, et qu'elle se rapproche à une vitesse de 0.038 m par an, on pourrait calculer l'évolution de la distance Terre-Lune suivant la formule :**

$$
d(t) = d_0 - 0.038 \times t \qquad (A.2)
$$

**avec $d_0$ la distance actuelle entre la Terre et la Lune, et $t$ le temps en années. En utilisant cette formule, à partir de quelle année la Lune franchirait la limite de Roche. (1.5 points)**

***Conseil** : Pour cette dernière question, utilisez `np.linspace` pour générer un tableau de temps allant jusqu'à 10 milliard d'années.*

````python
import numpy as np
import matplotlib.pyplot as plt

#1. 

def calcul_limite_roche(r_P, d_P, d_S):
    d_roche = 2.44 * r_P * (d_P / d_S)**(1/3)
    return d_roche

terre = {
    'nom': 'Terre',
    'densite': 5515,  # en kg/m^3
    'rayon': const.R_earth.value,  # en mètres
}

lune = {
    'nom': 'Lune',
    'densite': 3340,  # en kg/m^3
    'distance':3.84e8,  # en mètres
}

d_roche_terre = calcul_limite_roche(terre["rayon"], terre["densite"], lune["densite"])
print(f"La limite de Roche de la Terre est de {d_roche_terre} m.")


#2. 

def afficher_limite_roche(planete, satellite):

    rayon_planete = planete['rayon']
    densite_planete = planete['densite']
    densite_satellite = satellite['densite']
    distance_satellite = satellite['distance']

    d_roche = calcul_limite_roche(rayon_planete, densite_planete, densite_satellite)

    if  distance_satellite < 1.1*d_roche:
        print(f"{satellite['nom']} est proche de la limite de Roche de {planete['nom']}.")
    else:
        print(f"{satellite['nom']} est loin de la limite de Roche de {planete['nom']}.")

afficher_limite_roche(terre, lune)

#3.

jupiter = {
    'nom': 'Jupiter',
    'densite': 1328,  # en kg/m^3
    'rayon': const.R_jup.value,  # en mètres
}

io = {
    'nom': 'Io',
    'densite': 3528,  # en kg/m^3
    'distance': 4.22e8,  # en mètres
}

europe = {
    'nom': 'Europe',
    'densite': 3010,  # en kg/m^3
    'distance': 6.71e8,  # en mètres
}

adrastée = {
    'nom': 'Adrastée',
    'densite': 860,  # en kg/m^3 (hypothétique)
    'distance': 1.29e8,  # en mètres
}

satellites_jupiter = [io, europe, adrastée]


for satellite in satellites_jupiter:
    afficher_limite_roche(jupiter, satellite)

#4.

temps = np.linspace(0, 10*1e9)
v_lune = 3.8e-2
d_lune = lune["distance"] - v_lune*temps
d_roche_terre = calcul_limite_roche(terre["rayon"], terre["densite"], lune["densite"])

plt.figure()
plt.plot(temps, d_lune)
plt.axhline(y=d_roche_terre, color='r', linestyle='--', label='Limite de Roche actuelle de la Terre')
plt.legend()
plt.xlabel('Temps (années)')
plt.ylabel('Distance de la Lune à la Terre (m)')
plt.title('Évolution de la distance de la Lune à la Terre dans le futur')
plt.show()

limite_d_lune = d_lune < d_roche_terre
print(f"La Lune franchira la limite de Roche de la Terre dans {temps[limite_d_lune][0]/1e9} milliards d'années.")

````

# **Exercice 2.B — Simulation d'une mesure bruitée (7 points)**

La loi de **Stefan-Boltzman** permet de relier la luminosité d'une étoile à sa température de surface et à son rayon, telle que :
$$
L = 4\pi R^2 \sigma T^4 \qquad (B.1)
$$

avec :

- $L$ : luminosité de l'étoile (en watts).
- $R$ : rayon de l'étoile (en mètres).
- $\sigma$ : constante de Stefan-Boltzman ($5.67 \times 10^{-8}$ W/m²/K⁴).
- $T$ : température de surface de l'étoile (en kelvins).   

À l'aide d'un **télescope** nous avons observé une étoile et nous avons **mesuré** sa **luminosité** $L=3.828 \times 10^{26}$ W ainsi que sa **température** de surface $T = 5778$ K. En utilisant la loi de Stefan-Boltzman, nous pouvons déterminer le rayon de cette étoile en retournant l'équation telle que :

$$
R = \sqrt{\frac{L}{4\pi \sigma T^4}} \qquad (B.2)
$$

La température a été mesurée avec une incertitude négligeable, cependant la mesure de la luminosité possède une **incertitude** de $\sigma_L = 0.15 \times 10^{26}$ watts. L'incertitude sur la luminosité se traduit par une incertitude sur le rayon de l'étoile $\sigma_R$, que l'on peut calculer avec la formule :

$$
\sigma_R = \frac{1}{2} \sqrt{\frac{1}{4\pi \sigma T^4}} \times \frac{\sigma_L}{\sqrt{L}} \qquad (B.3)
$$  

**1. Calculez et affichez la valeur du rayon de l'étoile ainsi que son incertitude à l'aide des équations B.2 et B.3. (1 point)**

**2. Calculez le rapport de l'incertitude sur le rayon à la valeur du rayon. On considère que notre estimation de rayon est bonne si l'incertitude est inférieure à $1\%$ de la valeur, affichez si tel est le cas ici. (1 point)**

---

Lorsque l'on fait la moyenne du rayon $\bar{R}$ sur plusieurs mesures, on peut réduire l'incertitude de la mesure. En effet, si l'on fait la moyenne de $N$ mesures indépendantes, l'incertitude sur la moyenne $\sigma_{\bar{R}}$ est donnée par :

$$
\sigma_{\bar{R}} = \frac{\sigma_R}{\sqrt{N}} \qquad (B.4)
$$

**3. À partir de l'équation B.4, calculez le nombre de mesure $N$ nécessaire pour obtenir une incertitude $\sigma_{\bar{R}} < 0.01 \times R$. (1 point)**

**4. En simulant $N$ mesures de la luminosité de l'étoile $L$ à laquelle s'ajoute un bruit gaussien de moyenne $\mu_L = 0$ et d'écart-type $\sigma_L = 0.15 \times L$, vérifiez que la moyenne de ces mesures est proche de la valeur du rayon calculée précédemment. Vérifiez que l'incertitude sur la moyenne $\sigma_{\bar{R}}$ est bien de l'ordre de $0.01 \times R$. (2 points)**

````python
#1. 

import numpy as np
import astropy.constants as const

def calcul_rayon_etoile(luminosite, temperature):
    sigma = const.sigma_sb.value  # Constante de Stefan-Boltzmann en W/m^2/K^4
    rayon = np.sqrt(luminosite / (4 * np.pi * sigma * temperature**4))
    return rayon

def calcul_incertitude_rayon(luminosite, temperature, incertitude_luminosite):
    sigma = const.sigma_sb.value  # Constante de Stefan-Boltzmann en W/m^2/K^4
    incertitude_rayon = (1/2) * np.sqrt(1 /(4 * np.pi * sigma * temperature**4)) * (incertitude_luminosite / np.sqrt(luminosite)) 
    return incertitude_rayon

luminosite = 3.828e26  # en watts
temperature = 5778  # en Kelvin
incertitude_luminosite = 0.15*luminosite  # 10% d'incertitude sur la luminosité

rayon = calcul_rayon_etoile(luminosite, temperature)
print(f"Le rayon est d'environ {rayon} mètres.")

incertitude_rayon = calcul_incertitude_rayon(luminosite, temperature, incertitude_luminosite)
print(f"L'incertitude sur le rayon du Soleil est d'environ {incertitude_rayon} mètres.")

#2.

ratio = incertitude_rayon / rayon
print(f"Le ratio de l'incertitude sur le rayon par rapport au rayon lui-même est d'environ {ratio:.2%}.")


if ratio > 0.01:
    print("L'incertitude est supérieure à 0.01, l'estimation du rayon est incertaine.")
else:
    print("L'incertitude est inférieure à 0.01, l'estimation du rayon est précise.")

#3. 

def calcul_N_mesures(valeur, incertitude):
    N_mesures = (incertitude / (0.01*valeur))**2
    return N_mesures

N_mesures = calcul_N_mesures(rayon, incertitude_rayon)
print("Le nombre de mesures nécessaires pour réduire l'incertitude à 0.01 est d'environ", int(N_mesures))

#4. 
bruit_gaussien = np.random.normal(loc=0, scale=incertitude_luminosite, size=int(N_mesures))
luminosite_mesuree = luminosite + bruit_gaussien
rayon_mesure = calcul_rayon_etoile(luminosite_mesuree, temperature)

moyenne_rayon_mesure = np.mean(rayon_mesure)
incertitude_rayon_mesure = calcul_incertitude_rayon(luminosite, temperature, incertitude_luminosite/np.sqrt(N_mesures))
print(f"L'écart-type du rayon mesuré est d'environ {incertitude_rayon_mesure/moyenne_rayon_mesure:2%} mètres.")

````

````python
# 1.

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

stars = np.load('hyg_catalog.npy', allow_pickle=True)
temperatures = stars["temperature"]  # en K
luminosites = stars["lum"]   # en L_sun

plt.figure(figsize=(8,6))
plt.scatter(temperatures, luminosites, s=0.5, c=temperatures, cmap='plasma', norm=LogNorm())
plt.xscale('log')
plt.yscale('log')
plt.gca().invert_xaxis()
plt.xlabel('Température (K)')
plt.ylabel('Luminosité (L_sun)')
plt.title('Diagramme HR des étoiles du catalogue HYG')
plt.colorbar(label='Température (K)')

# 2.

soleil_temperature = 5778  # en K
soleil_luminosite = 1  # en L_sun

plt.scatter(soleil_temperature, soleil_luminosite, color='white', edgecolor='black', s=200, label='Soleil', marker="*")
plt.legend()
plt.show()

print("Le Soleil appartient à la catégorie des étoiles de la séquence principale.")

# 3.

mask_naines_blanches = (4000 < temperatures) & (luminosites < 0.005)
mask_geante_rouges = (temperatures < 5300) & (2.3 < luminosites) & (luminosites < 10e4)
mask_supergeantes = (luminosites > 10e4)
mask_sp = ~(mask_naines_blanches | mask_geante_rouges | mask_supergeantes)

# 4. 

L_moyenne_naines_blanches = np.mean(luminosites[mask_naines_blanches])
T_moyenne_naines_blanches = np.mean(temperatures[mask_naines_blanches])
print(f"Naines blanches : Luminosité moyenne = {L_moyenne_naines_blanches:.4f} L_sun, Température moyenne = {T_moyenne_naines_blanches:.2f} K")

L_moyenne_geante_rouges = np.mean(luminosites[mask_geante_rouges])
T_moyenne_geante_rouges = np.mean(temperatures[mask_geante_rouges])
print(f"Géantes rouges : Luminosité moyenne = {L_moyenne_geante_rouges:.4f} L_sun, Température moyenne = {T_moyenne_geante_rouges:.2f} K")

L_moyenne_supergeantes = np.mean(luminosites[mask_supergeantes])
T_moyenne_supergeantes = np.mean(temperatures[mask_supergeantes])
print(f"Supergéantes : Luminosité moyenne = {L_moyenne_supergeantes:.4f} L_sun, Température moyenne = {T_moyenne_supergeantes:.2f} K")  

L_moyenne_sp = np.mean(luminosites[mask_sp])
T_moyenne_sp = np.mean(temperatures[mask_sp])
print(f"Séquence principale : Luminosité moyenne = {L_moyenne_sp:.4f} L_sun, Température moyenne = {T_moyenne_sp:.2f} K")

# 5. 

import astropy.constants as const

def calcul_rayon_moyen(luminosite, temperature):
    sigma = const.sigma_sb.value  # Constante de Stefan-Boltzmann en W/m^2/K^4
    rayon = np.sqrt(luminosite * const.L_sun.value / (4 * np.pi * sigma * temperature**4))
    return rayon

rayon_naines_blanches = calcul_rayon_moyen(L_moyenne_naines_blanches, T_moyenne_naines_blanches)
print(f"Rayon moyen des naines blanches : {rayon_naines_blanches:.2e} m")
rayon_geante_rouges = calcul_rayon_moyen(L_moyenne_geante_rouges, T_moyenne_geante_rouges)
print(f"Rayon moyen des géantes rouges : {rayon_geante_rouges:.2e} m")
rayon_supergeantes = calcul_rayon_moyen(L_moyenne_supergeantes, T_moyenne_supergeantes)
print(f"Rayon moyen des supergéantes : {rayon_supergeantes:.2e} m")
rayon_sp = calcul_rayon_moyen(L_moyenne_sp, T_moyenne_sp)
print(f"Rayon moyen des étoiles de la séquence principale : {rayon_sp:.2e} m")  
````
