## **Exercice 6 — Calculer la luminosité d'une étoile avec la loi de Planck**

En utilisant les outils appris lors de cette séance, vous allez :

1. **Déterminer la luminosité totale** de votre étoile à l’aide de la **loi de Stefan-Boltzmann (<https://fr.wikipedia.org/wiki/Loi_de_Stefan-Boltzmann>)** :

   $$
   L = 4 \pi R^2 \sigma T^4
   $$

   où :
   - $L$ est la **luminosité totale** (puissance totale rayonnée) en $\mathrm{W}$,
   - $R$ est le **rayon de l’étoile** en $\mathrm{m}$,
   - $\sigma = 5.670374419 \times 10^{-8} \, \mathrm{W.m^{-2}.K^{-4}}$ est la **constante de Stefan-Boltzmann**,
   - $T$ est la **température de surface** de l’étoile, en $\mathrm{K}$.

2. **Calculez la longueur d'onde du pic d'émission à l'aide de la loi de Wien (<https://fr.wikipedia.org/wiki/Loi_du_d%C3%A9placement_de_Wien>)** : 

   $$
   \lambda_\mathrm{max} = \sigma_W / T
   $$

   où :
   - $\lambda_\mathrm{max}$ est la **longueur d'onde** du pic de luminosité en $\mathrm{m}$,
   - $\sigma_W = 2.898e-3 \, \mathrm{m.K^{-1}}$ est la **constante de Wien**,
   - $T$ est la **température de surface** de l’étoile, en $\mathrm{K}$.

3. **Calculer sa luminance spectrale** (ou intensité spectrale) à l’aide de la **loi de Planck (https://fr.wikipedia.org/wiki/Loi_de_Planck)** :

   $$
   B_\lambda(T) = \frac{2 h c^2}{\lambda^5} \cdot \frac{1}{e^{\frac{h c}{\lambda k_B T}} - 1}
   $$

   où :

   - $B_\lambda(T)$ est la **luminance spectrale** du corps noir, en $\mathrm{W.m^{-3}.sr^{-1}}$  
   - $\lambda$ est la **longueur d’onde** du rayonnement, en $\mathrm{m}$  
   - $T$ est la **température absolue** du corps noir, en $\mathrm{K}$  
   - $h = 6.62607015 \times 10^{-34} \, \mathrm{J.s}$ est la **constante de Planck**  
   - $c = 2.99792458 \times 10^8 \, \mathrm{m.s^{-1}}$ est la **vitesse de la lumière**  
   - $k_B = 1.380649 \times 10^{-23} \, \mathrm{J.K^{-1}}$ est la **constante de Boltzmann**

---

### Astuce

Il est possible d’utiliser le module **`astropy`** pour importer directement les constantes physiques fondamentales, sans avoir à les recopier manuellement :

```python
import astropy.constants as const

h = const.h.value       # constante de Planck (J·s)
c = const.c.value       # vitesse de la lumière (m/s)
k_B = const.k_B.value   # constante de Boltzmann (J/K)
sigma = const.sigma_sb.value  # constante de Stefan-Boltzmann (W/m²/K⁴)


Maintenant, à vous de **créer votre étoile** ! Pour vous inspirer, et vérifier que votre formule de luminosité rentre dans la gamme indiquée dans le tableau ci-dessous :

<figure style="text-align: center;">
  <img src="classification_etoile.png " alt="Tableau de classification des étoiles" />
  <figcaption>Tableau 1 : Classification spectrale des étoiles
    <a href="https://en.wikipedia.org/wiki/Stellar_classification#Harvard_spectral_classification" target="_blank">[https://en.wikipedia.org/wiki/Stellar_classification#Harvard_spectral_classification]</a>.
  </figcaption>
</figure>

````python
# Import des bibliothèques
import numpy as np
import matplotlib.pyplot as plt
import astropy.constants as const

# Constantes physiques
h = const.h.value            # constante de Planck (J·s)
c = const.c.value            # vitesse de la lumière (m/s)
k_B = const.k_B.value        # constante de Boltzmann (J/K)
sigma = const.sigma_sb.value # constante de Stefan-Boltzmann (W/m²/K⁴)
b = 2.898e-3                 # constante de Wien (m·K)

l_sol = const.L_sun.value          # luminosité solaire
m_sol = const.M_sun.value          # masse solaire
r_sol = const.R_sun.value          # raayon solaire

# Paramètres de l'étoile (à personnaliser)
nom_etoile = "Mon Étoile"
temperature = 10000          # température de surface (K)
radius = 1.8*r_sol          # rayon (m) ~ rayon solaire
mass = 2.1*m_sol             # masse (kg) ~ masse solaire

# Calcul de la luminosité (L = 4πR²σT⁴)
def stefan_law(T, R):
    """Calcule la luminosité totale d'un corps noir."""
    return 4 * np.pi * R**2 * sigma * T**4

# Calcul du pic d'émission : λ_max = b / T (b = 2.898e-3 m·K)
def wien_law(T):
    """Calcule la longueur d'onde du maximum d'émission."""
    return b / T

# Affichage des propriétés (Nom de l'étoile, masse, rayon, température, luminosité, longueur d'onde du pic d'émission)
luminosity = stefan_law(temperature, radius)
lambda_max = wien_law(temperature)

sun_symbol = "\u2609"
print(f"Propriétés de l'étoile : {nom_etoile}")
print(f"Température : {temperature:.0f} K")
print(f"Rayon       : {radius:.2e} m")
print(f"Masse       : {mass:.2e} kg")
print(f"Luminosité  : {luminosity:.2e} W = {luminosity/l_sol:.2e} L{sun_symbol}")
print(f"λ_max (Wien): {lambda_max*1e9:.1f} nm")

# Spectre du corps noir (simplifié)
def planck_law(lam, T):
    """
    Renvoie la luminance spectrale B_lambda(T)
    lam : longueur d'onde en m
    T   : température en K
    """
    return (2 * h * c**2) / (lam**5 * (np.exp((h * c) / (lam * k_B * T)) - 1))

# Plage de longueurs d'onde (en mètres)
wavelengths = np.linspace(100e-9, 3000e-9, 1000)

# Calcul du spectre
B_lambda = planck_law(wavelengths, temperature)

# Visualisation
plt.figure(figsize=(8,5))
plt.plot(wavelengths*1e9, B_lambda)
plt.title(f"Spectre du corps noir – {nom_etoile} (T = {temperature} K)")
plt.xlabel("Longueur d'onde λ (nm)")
plt.ylabel(r"Luminance spectrale $B_\lambda(T)$ (W·m$^{-3}$·sr$^{-1}$)")
plt.grid(True)
plt.axvline(lambda_max*1e9, color='r', linestyle='--', label=f"λ_max = {lambda_max*1e9:.0f} nm")
plt.legend()
plt.tight_layout()
plt.show()
````

## **Exercice 7 — Sauvegarder les tableaux puis les charger**

Le package numpy permet de sauvegarder des **`np.array()`** en utilisant la fonction **`np.save()`**, qu'il est possible de charger ensuite avec la fonction **`np.load()`**. Répétez l'affichage et le plot précédent pour vous assurez d'avoir correctement sauvegardé et chargé les données.

## **Pour aller plus loin**

Il est possible de produire des graphiques interactifs avec la librairie **`ipywidgets`**. En utilisant les widgets tels que les sliders, il est possible de modifier les paramètres d'une fonction et de visualiser en temps réel les changements sur le graphique. Voici un exemple d'implémentation pour la loi de Planck, où il est possible de modifier la température de l'étoile.

````python
from ipywidgets import interact, FloatSlider, IntSlider, VBox

def interactive_planck_plot(temperature):
    wavelengths = np.linspace(100e-9, 5000e-9, 1000)
    B_lambda = planck_law(wavelengths, temperature)

    plt.figure(figsize=(8, 4))
    plt.plot(wavelengths, B_lambda, color='darkorange', label="Luminance")
    plt.legend()    
    plt.grid(True)
    plt.title(f"Spectre du corps noir – {nom_etoile} (T = {temperature} K)")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

temperature_slider=FloatSlider(value=1, min=2000, max=10000, step=500, description="temperature")
ui = VBox([temperature_slider]) 

interact(interactive_planck_plot, temperature=temperature_slider) 
````

````python
from ipywidgets import interact, FloatSlider, IntSlider, Checkbox, VBox

# Constantes physiques
h = const.h.value            # constante de Planck (J·s)
c = const.c.value            # vitesse de la lumière (m/s)
k_B = const.k_B.value        # constante de Boltzmann (J/K)

# Loi de Planck
def planck_law(lam, T):
    """
    Renvoie la luminance spectrale B_lambda(T)
    lam : longueur d'onde en m
    T   : température en K
    """
    return (2 * h * c**2) / (lam**5 * (np.exp((h * c) / (lam * k_B * T)) - 1))

# Fonction interactive
def interactive_planck_plot(temperature, wavelength_min, wavelength_max, pic_emission):
    wavelengths = np.linspace(wavelength_min * 1e-9, wavelength_max * 1e-9, 1000)
    B_lambda = planck_law(wavelengths, temperature)

    plt.figure(figsize=(10, 6))
    plt.plot(wavelengths * 1e9, B_lambda, color='darkorange', label="Luminance")
    plt.title(f"Spectre du corps noir – T = {temperature} K")
    plt.xlabel("Longueur d'onde λ (nm)")
    plt.ylabel(r"Luminance spectrale $B_\lambda(T)$ (W·m$^{-3}$·sr$^{-1}$)")
    plt.grid(True)
    if pic_emission:
        plt.axvline((2.898e-3 / temperature) * 1e9, color='red', linestyle='--', label=r"$\lambda_{max}$ (Loi de Wien)")
    plt.legend()
    plt.show()

# Widgets interactifs
temperature_slider = FloatSlider(value=5000, min=2000, max=10000, step=500, description="Température (K)")
wavelength_min_slider = IntSlider(value=100, min=10, max=990, step=10, description="λ min (nm)")
wavelength_max_slider = IntSlider(value=3000, min=1000, max=10000, step=100, description="λ max (nm)")
pic_emission_checkbox = Checkbox(value=False, description="Affichage pic d'émission")

# Interface utilisateur
ui = VBox([temperature_slider, wavelength_min_slider, wavelength_max_slider, pic_emission_checkbox])
interact(interactive_planck_plot, 
         temperature=temperature_slider, 
         wavelength_min=wavelength_min_slider, 
         wavelength_max=wavelength_max_slider, 
         pic_emission=pic_emission_checkbox)
````
